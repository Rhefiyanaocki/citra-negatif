```python
# import library

from PIL import Image
```

```python
# load citra

CITRA = Image.open('hii.png')
```

```python
ukuran_horizontal = CITRA.size[0]
ukuran_vertikal = CITRA.size[1]
```

```python
PIXEL = CITRA.load()
```

```python
ukuran_horizontal
```

```python
ukuran_vertikal
```

```python
for x in range(ukuran_horizontal):
    for y in range(ukuran_vertikal):
        R = 150 - PIXEL[x, y][0]
        G = 150 - PIXEL[x, y][1]
        B = 150 - PIXEL[x, y][2]
        PIXEL[x, y] = (R, G, B)
```

```python
CITRA.save('gambar_negatif.png')
```
